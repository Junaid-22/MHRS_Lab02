```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
img1 = rasterio.open("D:\Sir Ehtsham\Python\Extract.TIF", "r")
Band_3 = img1.read(3).astype('float32')
Band_6= img1.read(6).astype('float32')
ndsi_clipped = (Band_3 - Band_6) / (Band_3 + Band_6)
ndsi_clipped = np.nan_to_num(ndsi_clipped, nan=-1)
plt.imshow(ndsi_clipped, cmap='gray')
plt.colorbar()
plt.show()
```
